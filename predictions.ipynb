```python
predictors = ["interest", "vacancy", "adj_price", "adj_value"]
target = "change"
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

Start with 5 years of data, 2008-2013, then make predictions for next year (2014), then 2008-2014 and make predictions for 2015. 
Split data, training is everything up until i, test is year following i. Step is 52 as 52 weeks in the year. Concanate predictions into an array, return preditions and accuracy. 


#min_sample, prevents overfitting
#initialize model, fit on training data and predict on testing data.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np

START = 260
STEP = 52

def predict(train, test, predictors, target):
    rf = RandomForestClassifier(min_samples_split=10, random_state=1)
    rf.fit(train[predictors], train[target])
    preds = rf.predict(test[predictors])
    return preds

def backtest(data, predictors, target):
    all_preds = []
    for i in range(START, data.shape[0], STEP):
        train = price_data.iloc[:i]
        test = price_data.iloc[i:(i+STEP)]
        all_preds.append(predict(train, test, predictors, target))
    
    preds = np.concatenate(all_preds)
    return preds, accuracy_score(data.iloc[START:][target], preds)
```

```python
preds, accuracy = backtest(price_data, predictors, target)
```

Continue improving 

Give recent trend to improve , using pandas rolling method. Give model extra information about recent trend. 

Ratio between current sales price, interst rates and average during the past year. 

Use pandas rolling method to pass in minimum period, so even if it is 1 period of data there will not be any NA. 

```python
#Gives one period of data
#Gives averages for the past year. 
yearly = price_data.rolling(52, min_periods=1).mean()
```

```python
yearly
```

Find ratio between current value and value in the past year. Taking interest, price ect and dividing by average. I.e. if 1.46 interest rate, it shows interest rates is much higher than they were over the last year. Gives the algo more information. 

```python
yearly_ratios = [p + "_year" for p in predictors]
price_data[yearly_ratios] = price_data[predictors] / yearly[predictors]
```

```python
price_data
```

```python
preds, accuracy = backtest(price_data, predictors + yearly_ratios, target)
```

```python
accuracy
```
